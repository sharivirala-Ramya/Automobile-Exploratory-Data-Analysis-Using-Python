<h1>6.Data aggregation and pivot</h1>

```python
df["drive-wheels"].unique()
```

```python
#we can select the columns drive-wheels ,body-style and price then assign it to the variable group df_group_one


df_group_one = df[['drive-wheels','body-style','price']]
df_group_one
```

```python
#we can then calculate the avg price for each of the diff categories of data

df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean(numeric_only=True)
df_group_one

```

```python
df_gptest = df[['drive-wheels','body-style','price']]
df_gptest = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean(numeric_only=True)
df_gptest 
```

```python
A better way to represent above data is using pivot which is used in excel
```

```python
grouped_pivot = df_gptest.pivot(index ='drive-wheels',columns = 'body-style')
grouped_pivot
```

```python
Often we wont have data for some of the pivot cells.we can fill these missing cells with the value 0,but any other value could potentiallu be used as well
```

```python
grouped_pivot=grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```
